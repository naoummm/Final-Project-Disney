```python
# Load the dataset
data = pd.read_csv("../input/disney-movies-and-tv-shows/disney_plus_titles.csv")

# Take user input
search_query = input("Enter your search query: ")

# Create a lowercase column for comparison
data['lower_title'] = data['title'].str.lower()
data['lower_description'] = data['description'].str.lower()
data['lower_listed_in'] = data['listed_in'].str.lower()

# Filter the data based on user input
filtered_data = data[
    (data['lower_title'].str.contains(search_query.lower())) |
    (data['lower_description'].str.contains(search_query.lower())) |
    (data['lower_listed_in'].str.contains(search_query.lower()))
].copy()

# Rename the 'listed_in' column to 'categories'
filtered_data.rename(columns={'listed_in': 'categories'}, inplace=True)

# Display the top 10 matching shows
filtered_data[['title', 'description', 'categories']]
```

```python

# Load the dataset
df_strategy = pd.read_csv("../input/disney-movies-and-tv-shows/disney_plus_titles.csv")

# Convert 'release_year' column to numeric
df_strategy['release_year'] = pd.to_numeric(df_strategy['release_year'], errors='coerce')

# Filter the data for the year 2013 and onwards
df_filtered = df_strategy[df_strategy['release_year'] >= 2011]

# Count the number of TV shows and movies for each year
counts = df_filtered.groupby(['release_year', 'type']).size().unstack().fillna(0)

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(counts, annot=True, fmt='g', cmap='YlGnBu')
plt.title("Disney concentration: TV Shows vs. Movies ")
plt.xlabel("Type")
plt.ylabel("Year")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'type' and calculate the size
grouped_data = df_filtered.groupby(['type']).size()

# Create the pie chart
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%')

# Add title
plt.title('TV Shows vs. Movies between 2011-2021')

# Show the plot
plt.show()

```
